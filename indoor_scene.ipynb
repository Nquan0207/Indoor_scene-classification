```python
from pathlib import Path
import re

dataset_dir = Path("data/indoorCVPR_09/Images")

def sanitize_name(name: str) -> str:
    sanitized = re.sub(r"[^A-Za-z0-9._-]", "_", name)
    sanitized = re.sub(r"_+", "_", sanitized)
    return sanitized

renamed = []
for path in dataset_dir.rglob("*"):
    if not path.is_file():
        continue

    new_name = sanitize_name(path.name)
    if new_name == path.name:
        continue

    new_path = path.with_name(new_name)
    counter = 1
    while new_path.exists():
        stem = Path(new_name).stem
        suffix = Path(new_name).suffix
        new_path = path.with_name(f"{stem}_{counter}{suffix}")
        counter += 1

    path.rename(new_path)
    renamed.append((path.name, new_path.name))

print(f"Renamed {len(renamed)} files")
for old_name, new_name in renamed[:20]:
    print(f"{old_name} -> {new_name}")
```
